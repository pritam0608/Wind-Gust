```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
df = pd.read_csv("D:\\Summer training\\Met_data_1.csv")
```

```python
df['WIND_MAX/GUST_10m(km/hr)'] = df['WIND_MAX/GUST_10m(kt)'] * 1.852

df['WIND_SPEED_10m(km/hr)'] = df['WIND_SPEED_10m(kt)'] * 1.852

```

```python
# Round wind speed values to two decimal places
df['WIND_MAX/GUST_10m(km/hr)'] = df['WIND_MAX/GUST_10m(km/hr)'].round(2)

df['WIND_SPEED_10m(km/hr)'] = df['WIND_SPEED_10m(km/hr)'].round(2)

```

```python
# Save the updated dataset to a new CSV file
df.to_csv('updated_Met_data_1.csv', index=False)
```
